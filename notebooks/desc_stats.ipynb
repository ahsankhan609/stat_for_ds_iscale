## There are Two main types of Statistics:
1. Descriptive Statistics
2. Inferential Statistics

### Descriptive Statistics
Descriptive statistics is a branch of statistics that focuses on summarizing and describing the main features of a dataset. It involves the use of various measures and techniques to provide a clear and concise overview of the data. Descriptive statistics helps in understanding the central tendency, variability, and distribution of the data, making it easier to interpret and communicate findings. Some common measures used in descriptive statistics include:
- Mean: The average value of a dataset.
- Median: The middle value when the data is ordered.
- Mode: The most frequently occurring value in a dataset.
- Standard Deviation: A measure of the amount of variation or dispersion in a dataset.
- Range: The difference between the maximum and minimum values in a dataset.
- Variance: The average of the squared differences from the mean, indicating how much the data points deviate from the mean.
- Percentiles: Values that divide the data into 100 equal parts, indicating the relative standing of a particular value within the dataset.
Descriptive statistics is often used in the initial stages of data analysis to summarize and explore the data before applying more complex statistical methods or making inferences about the population. It provides a foundation for understanding the data and can help identify patterns, trends, and outliers that may be important for further analysis.
### Inferential Statistics
Inferential statistics is a branch of statistics that allows us to make conclusions and predictions about a population based on a sample of data. It involves using probability theory and statistical methods to draw inferences about the population parameters, such as the mean, proportion, or variance, based on the sample data. Inferential statistics helps us to generalize our findings from the sample to the larger population and to test hypotheses about the relationships between variables. Some common techniques used in inferential statistics include:
- Hypothesis Testing: A method for testing a claim or hypothesis about a population parameter based on sample data.
- Confidence Intervals: A range of values that is likely to contain the true population parameter with a certain level of confidence.
- Regression Analysis: A statistical technique used to model the relationship between a dependent variable and one or more independent variables.
- Analysis of Variance (ANOVA): A method for comparing the means of three or more groups to determine if there are significant differences between them.
- Chi-Square Test: A statistical test used to determine if there is a significant association between two categorical variables.
Inferential statistics is essential for making informed decisions and drawing conclusions about a population based on limited data. It allows us to assess the reliability of our findings and to make predictions about future outcomes based on the observed data. Inferential statistics is widely used in various fields, including social sciences, medicine, economics, and business, to analyze data and make evidence-based decisions.

## Calculating Mean with different Methods - Manual, Numpy, Builtin Statistics

#### Manual Mean calculation

```python
population: list[int] = [150, 160, 170, 165, 155]  # population data
sample_data: list[int] = [150, 170, 165]

# calculating population mean
mean_population: float = sum(population) / len(population)
mean_sample: float = sum(sample_data) / \
    len(sample_data)  # calculating sample mean

print(mean_population)
print(mean_sample)
```

#### Numpy Mean calculation

```python
import numpy as np

print("Populaiton Mean:", np.mean(population))
print("Sample Mean:", np.mean(sample_data))
```

#### Builtin Statistics Mean calculation

```python
import statistics as stats

print("Populaiton Mean: using builtin statisctics method", stats.mean(population))
print("Sample Mean: using builtin statistics method", stats.mean(sample_data))
```

## Calculating Median with different Methods - Numpy, Builtin Statistics and Pandas

```python
import numpy as np
import pandas as pd
import statistics as stats

data: list[int] = [10, 20, 30, 40, 50, 60, 70, 80, 90, 150]
df = pd.DataFrame({
    'values': [10, 20, 30, 40, 50, 60, 70, 80, 90, 150]
})
print("Median with Numpy is:", np.median(data))
print("Median with builtin Statistics is:", stats.median(data))
medain_with_pandas = df['values'].median()
print(medain_with_pandas)
```

## Calculating Mode, Multi Mode with different Methods - Pandas, statistics and Scipy

```python
# import numpy as np
import pandas as pd
import statistics as stats
from scipy import stats as sci_stat

data: list[int] = [10, 10, 10, 20, 20, 20,
                   30, 40, 50, 60, 60, 60, 70, 80, 90, 150]
df = pd.DataFrame(
    {'data_value': [10, 20, 30, 40, 50, 60, 70, 80, 90, 150, 165]})
print("Mode with Pandas is:", df['data_value'].mode())

print("Mode with builtin Statistics is:", stats.mode(data))
# statistics always shows the Mode(number)that is smallest.
# builtin stats module only returns first small number, if there are multiple modes.
# mode is the value that appears most frequently in a dataset. For a dataset to be multimodal, two or more values must have the exact same highest frequency.
print("mode is the value that appears most frequently in a dataset.\n For a dataset to be multimodal, two or more values must have the exact same highest frequency.\n Multi Mode with builtin Statistics is:", stats.multimode(data))

print("Mode with the SciPy library using Stats module", sci_stat.mode(data))
```

## Variance Calculation - Numpy, Pandas, Scipy

```python
pop_data: list[int] = [2, 4, 4, 5, 5, 5, 6, 6, 6, 7, 7, 7, 7, 8, 8, 9]
sam_data: list[int] = [2, 4, 5, 6, 7, 8, 9]


# delta degree of freedom(by default set as 0 for population variance
print("Population Variance with Numpy:", np.var(pop_data, ddof=0))
# delta degree of freedom(set 1 for sample variance)
print("Sample Variance with Numpy:", np.var(sam_data, ddof=1))

print("\nfinding variance wiht the help of pandas")
pandas_series_for_popdata = pd.Series(pop_data)
pandas_series_for_samdata = pd.Series(sam_data)

print("Population Variance with Pandas:",
      pandas_series_for_popdata.var(ddof=0))
print("Sample Variance with Pandas:", pandas_series_for_samdata.var())

print("\nfinding variance wiht the help of Scipy")
print("Population Variance with Scipy:", sci_stat.tvar(pop_data, ddof=0))
print("Sample Variance with Scipy:", sci_stat.tvar(sam_data, ddof=1))
```

## Data Representation

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create a figure and axes

fig, axs = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle('Histogram with KDE of variuos Distributions', fontsize=16)

#  Normal- Guassain Distribution
normal_data = np.random.normal(loc=0,  scale=1, size=1000)    # mean=0, std =1
sns.histplot(normal_data, kde=True,

             ax=axs[0, 0], color='blue', stat='density', bins=30)
axs[0,  0].set_title('Normal Distribution')
axs[0, 0].grid(True)

# Skewed Distribution
data_right_skewed = np.random.exponential(scale=1, size=1000)
data_left_skewed = 5 - np.random.exponential(scale=1, size=1000)

# Right Skewed Distribution
sns.histplot(data_right_skewed, kde=True,
             ax=axs[0, 1], color='red', alpha=0.5, stat='density', bins=30, label='Right Skewed')
# Left Skewed Distribution
sns.histplot(data_left_skewed, kde=True,
             ax=axs[0, 1], color='green', alpha=0.5, stat='density', bins=30, label='Left Skewed')
axs[0,  1].set_title('Skewed Distribution')
axs[0, 1].legend()
axs[0, 1].grid(True)

#  Uniform Distribution
data_uniform = np.random.uniform(low=0, high=1,  size=1000)
sns.histplot(data_uniform, kde=True,

             ax=axs[1, 0], color='purple', stat='density', bins=30)
axs[1,  0].set_title('Uniform Distribution')
axs[1, 0].grid(True)

# Bimodal Distribution
data_bimodal = np.concatenate([np.random.normal(
    loc=-2, scale=1, size=500), np.random.normal(loc=2, scale=1, size=500)])
sns.histplot(data_bimodal, kde=True,

             ax=axs[1, 1], color='orange', stat='density', bins=30)
axs[1,  1].set_title('BiModal Distribution')
axs[1, 1].grid(True)

# MultiModal Distribution
data_multimodal = np.concatenate([
    np.random.normal(loc=-6, scale=1.5, size=500),
    np.random.normal(loc=0, scale=1, size=500),
    np.random.normal(loc=6, scale=1.5, size=500)
])

sns.histplot(data_multimodal, kde=True,

             ax=axs[2, 0], color='brown', stat='density', bins=30)
axs[2,  0].set_title('MultiModal Distribution')
axs[2, 0].grid(True)

#  hide empty subplot
axs[2, 1].axis('off')
#  adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
